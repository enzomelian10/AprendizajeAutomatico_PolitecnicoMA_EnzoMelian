```python
import pandas as pd

print('1. Cargar los datos de ventas desde el archivo CSV y mostrar una vista previa de las primeras filas.\n')
ventas_csv = pd.read_csv('ventas.csv', encoding = "ISO-8859-1", nrows=2)
print(ventas_csv)

print('\n2. Cargar los datos de clientes desde el archivo JSON y mostrar la cantidad total de clientes.\n')
clientes_json = pd.read_json('clientes.json')
total_clientes = clientes_json['id_cliente'].count()
print('La cantidad total de clientes es: ', total_clientes)

print('\n3. Cargar los datos de inventario desde la hoja de cálculo Excel y calcular el promedio de stock disponible para todos los productos.\n')
inventario_excel = pd.read_excel('inventario.xlsx')
stock_inventario_excel = inventario_excel['stock'].mean()
print('El stock en promedio disponible para todos los productos es: ', int(stock_inventario_excel))
```
