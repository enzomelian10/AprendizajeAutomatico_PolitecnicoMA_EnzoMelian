```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

#1. Cargar el conjunto de datos en una estructura adecuada para su manipulaciÃ³n.
automobile_csv = pd.read_csv('automobile.csv', encoding='ISO-8859-1')

#Visualizacion de tablas
automobile_csv.head()
```

```python
#Visualizacion de info de la tabla
automobile_csv.info()
```

```python
#2. Crear un grÃ¡fico de dispersiÃ³n para visualizar las relaciones entre los pares de caracterÃ­sticas numÃ©ricas `mpg` y `weight`.
plt.scatter(automobile_csv['mpg'],automobile_csv['weight'])

plt.xlabel('Miles per gallon "MPG" (Millas por galÃ³n)')
plt.ylabel('Weight (Peso)')
plt.title("RelaciÃ³n entre MPG y Weight")
plt.grid(True)

plt.show()
```

```python
#3. Generar un histograma del consumo de combustible (`mpg`) para analizar su distribuciÃ³n.
plt.hist(automobile_csv['mpg'], bins=15, rwidth=0.9)

plt.xlabel('MPG (millas por galon)')
plt.ylabel('Quantity (cantidad)')
plt.title('Histograma del consumo de combustible (mpg)')
plt.grid(True)

plt.show()
```

```python
#4. Graficar un diagrama de caja (boxplot) de las caracterÃ­sticas `cylinders` y `origin` para identificar posibles valores atÃ­picos
automobile_csv.boxplot(column='cylinders', by='origin')

plt.grid(False)
plt.title('cylinders (agrup. por origen de cilindros)')
plt.show()
```

```python
#5. Realizar un grÃ¡fico de barras que muestre la distribuciÃ³n de automÃ³viles segÃºn su aÃ±o de modelo (`model_year`) y origen (`origin`).
plt.figure(figsize=(12, 6))
sns.countplot(data=automobile_csv, x='model_year', hue='origin')

plt.title('Cantidad de autos por aÃ±o y origen')
plt.xlabel('AÃ±o del modelo')
plt.ylabel('Cantidad de autos')

plt.show()
```
